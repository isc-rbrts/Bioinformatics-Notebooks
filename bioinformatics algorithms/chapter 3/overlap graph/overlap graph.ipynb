```python
from collections import defaultdict

f = open("dataset_865398_10.txt","r")

Patterns = [str(x) for x in next(f).split()]

def OverlapGraph(Patterns):
    graph = defaultdict(list)
    for i in range(len(Patterns)):
        for j in range(len(Patterns)):
            if i != j and Patterns[i][:len(Patterns[i]) - 1] == Patterns[j][1:]:
                graph[Patterns[j]].append(Patterns[i])
                
    return graph

g = OverlapGraph(Patterns)
for key in g:
    out = key + ": "
    for node in g[key]:
        out += node + " "
    print(out)

f.close()
```

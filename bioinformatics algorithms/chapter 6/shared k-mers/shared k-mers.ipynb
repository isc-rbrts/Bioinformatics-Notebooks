```python
from collections import defaultdict

f = open("dataset_865471_5.txt","r")
k = int(f.readline())
s, t = f.readline().strip('\n'), f.readline().strip('\n')
f.close()

def SharedKMers(k, s, t):
    s_freqs = defaultdict(list)
    t_freqs = defaultdict(list)
    
    shared_kmers = set()
    
    for i in range(len(s) - k + 1):
        kmer = s[i:i+k]
        s_freqs[kmer].append(i)
        s_freqs[ReverseComplement(kmer)].append(i)
        
    for i in range(len(t) - k + 1):
        kmer = t[i:i+k]
        t_freqs[kmer].append(i)
        t_freqs[ReverseComplement(kmer)].append(i)
        
    for kmer in s_freqs:
        for index_s in s_freqs[kmer]:
            for index_t in t_freqs[kmer]:
                shared_kmers.add((index_s, index_t))
            
    return shared_kmers
        
def ReverseComplement(pattern):
    complement = ""
    complements = {'A':'T', 'T':'A', 'C':'G', 'G':'C'}
    
    for c in pattern:
        complement = complements[c] + complement
    return complement

for pair in SharedKMers(k, s, t):
    print("(" + str(pair[0]) + ", " + str(pair[1]) + ")")
```

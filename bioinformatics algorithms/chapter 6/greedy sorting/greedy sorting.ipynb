```python
from copy import deepcopy

f = open("dataset_865464_4.txt","r")
P = [int(x) for x in next(f).split()]
f.close()

def GreedySorting(P):
    permutations = []
    approxRevDist = 0
    
    #check each k
    for k in range(1, len(P) + 1):
        
        #if element k is not k
        if P[k-1] != k and P[k-1] != -k:
            
            #look through upper elements
            for i in range(k+1, len(P) + 1):
                
                #save current element at position k
                temp = P[k-1]
                
                #found upper element
                if P[i-1] == k or P[i-1] == -k:
                    
                    #reverse elements between found and k
                    c = deepcopy(P[k-1:i])
                    for j in range((i) - (k-1)):
                        P[j + (k-1)] = -c[-(j+1)]
                    permutations.append(deepcopy(P))
                    approxRevDist += 1
        
        #if the kth element is negative, reverse itself
        if P[k-1] == -k:
            P[k-1] = k
            permutations.append(deepcopy(P))
            approxRevDist += 1
        
    return permutations

for item in GreedySorting(P):
    out = ""
    for i in item:
        if i > 0:
            out += "+" + str(i) + " "
        else:
            out += str(i) + " "
    print(out.strip(" "))
```

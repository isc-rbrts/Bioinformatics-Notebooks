```python
from collections import defaultdict
from copy import deepcopy

f = open("dataset_865469_5.txt","r")
p, q = f.readline().split(')'), f.readline().split(')')
f.close()
P, Q = [], []
for chromosome in p[:-1]:
    P.append([int(x) for x in chromosome[1:].split()])
for chromosome in q[:-1]:
    Q.append([int(x) for x in chromosome[1:].split()])
    
def TwoBreakSorting(P, Q):
    Genomes = [P]
    BlueEdges = ColoredEdges(Q)
    #print(BlueEdges)
    max_cycles = len(Cycles(BreakPointGraph(Q, Q)))
    cycles = Cycles(BreakPointGraph(P, Q))
    while len(cycles) < max_cycles:
        cycle = []
        for c in cycles:
            if len(c) > 2:
                cycle = c
                break
        
        #print("cycle " + str(cycle) + " " + str(len(cycles)))
        no_match_flag = 1
        for i in range(1, len(cycle) - 1, 2):
                if (cycle[i], cycle[i+1]) in BlueEdges:
                    no_match_flag = 0
                    P = TwoBreakOnGenome(P, cycle[i-1], cycle[i], cycle[i+2], cycle[i+1], BlueEdges)
                    break
        if no_match_flag and (cycle[0], cycle[-1]) in BlueEdges:
            P = TwoBreakOnGenome(P, cycle[-2], cycle[-1], cycle[1], cycle[0], BlueEdges)
            
        Genomes.append(P)
        cycles = Cycles(BreakPointGraph(P, Q))
      
    return Genomes

def TwoBreakOnGenome(P, i1, i2 , i3, i4, BlueEdges):
    GenomeGraph = ColoredEdges(P)
    #print("input " + str(GenomeGraph))
    GenomeGraph = TwoBreakOnGenomeGraph(GenomeGraph, i1, i2, i3, i4)
    #print("two break " + str(GenomeGraph))
    GenomeGraph = FixGenomeGraph(GenomeGraph, BlueEdges)
    #print("fix " + str(GenomeGraph))
    p = GraphToGenome(GenomeGraph, BlueEdges)
    #print("p " + str(p) + "\n")
    return p

def FixGenomeGraph(graphP, graphQ):
    edges = graphP + graphQ
    
    graph = defaultdict(list)
    for edge in edges:
        graph[edge[0]].append(edge[1])
        graph[edge[1]].append(edge[0])
        
    cycles = Cycles(graph)
    
    newGraph = []
    for cycle in cycles:
        for i in range (0, len(cycle), 2):
            newGraph.append((cycle[i], cycle[i+1]))
            
    return newGraph

def TwoBreakOnGenomeGraph(GenomeGraph, i1, i2, i3, i4):
    order_flag = 0
    index_i1 = 0
    index_i3 = 0
    
    for edge in range(len(GenomeGraph)):
        if GenomeGraph[edge] == (i1, i2) or GenomeGraph[edge] == (i2, i1):
            if GenomeGraph[edge] == (i2, i1):
                order_flag = 1
            index_i1 = edge
        elif GenomeGraph[edge] == (i3, i4) or GenomeGraph[edge] == (i4, i3):
            index_i3 = edge
    
    if index_i3 < index_i1:
        if not order_flag:
            #print("here 1")
            return GenomeGraph[:index_i3] + [(i4, i2)] + GenomeGraph[index_i1 +1:] + \
                                                [(i1, i3)] + GenomeGraph[index_i3+1:index_i1]
        else:
            #print("here 2")
            return [(i2, i4)] + GenomeGraph[:index_i3] + [(i1, i3)] + GenomeGraph[index_i1 +1:] + \
                                                 GenomeGraph[index_i3+1:index_i1]
    else:
        if not order_flag:
            #print("here 3")
            return GenomeGraph[:index_i1] + [(i1, i3)] + GenomeGraph[index_i3 +1:] + \
                                                [(i2, i4)] + GenomeGraph[index_i1+1:index_i3]
        else:
            #print("here 4")
            return GenomeGraph[:index_i1] + [(i2, i4)] + GenomeGraph[index_i3 +1:] + \
                                                [(i1, i3)] + GenomeGraph[index_i1+1:index_i3]
    
    
def BreakPointGraph(P, Q):
    edges = ColoredEdges(P) + ColoredEdges(Q)
    
    graph = defaultdict(list)
    for edge in edges:
        graph[edge[0]].append(edge[1])
        graph[edge[1]].append(edge[0])
    
    return graph

def Cycles(graph):
    cycles_count = 0
    cycles = []
    
    unused_vertices = []
    for node in graph:
        unused_vertices.append(node)
    
    while len(unused_vertices):
        node = unused_vertices[0]
        cycle = [node]
        
        unused_vertices.remove(node)
            
        curr = graph[node][0]
        graph[node].remove(curr)
        graph[curr].remove(node)
        
        while curr != node:
            cycle.append(curr)
            
            unused_vertices.remove(curr)
                
            nxt = graph[curr][0]
            graph[curr].remove(nxt)
            graph[nxt].remove(curr)
            curr = nxt
        cycles_count += 1
        cycles.append(cycle)
            
    return cycles

def ColoredEdges(P):
    Edges = []
    for chromosome in P:
        Nodes = ChromosomeToCycle(chromosome)
        for i in range(1,len(Nodes) - 2, 2):
            Edges.append((Nodes[i], Nodes[i + 1]))
            
        Edges.append((Nodes[-1], Nodes[0]))
    return Edges
    
def ChromosomeToCycle(chromosome):
    Nodes = []
    for i in chromosome:
        if i > 0:
            Nodes.append(2*i - 1)
            Nodes.append(2*i)
        else:
            Nodes.append(-2*i)
            Nodes.append(-2*i - 1)
            
    return Nodes

def GraphToGenome(GenomeGraph, BlueEdges):
    P = []
    #print(GenomeGraph, BlueEdges)
    
    graph = dict()
    c_graph = dict()
    
    unused_vertices = []
    for edge in GenomeGraph:
        unused_vertices.append(edge[0])
        unused_vertices.append(edge[1])
    
    for i in range(1, len(GenomeGraph) * 2, 2):
        c_graph.update({i:i+1})
        c_graph.update({i+1:i})   
    
    for edge in GenomeGraph:
        graph.update({edge[0]: edge[1]})
        graph.update({edge[1]: edge[0]})

    while len(unused_vertices):
        chromosome = []
        
        start = unused_vertices[0]
        unused_vertices.remove(start)
        if start % 2 == 1:
            chromosome.append(-int((start+1)/2))
        else:
            chromosome.append(int((start)/2))
        
        curr = graph[start]
        #print("start: " + str(start) + " -> " + str(curr))
        unused_vertices.remove(curr)
        #print(str(curr) + " -> " + str(c_graph[curr]))
        curr = c_graph[curr]
        
        while curr != start:
            unused_vertices.remove(curr)
            #print("checking " + str(curr))
            if curr % 2 == 1:
                chromosome.append(-int((curr+1)/2))
            else:
                chromosome.append(int((curr)/2))
                
            #print(str(curr) + " -> " + str(graph[curr]))
            curr = graph[curr]
            unused_vertices.remove(curr)
            #print(str(curr) + " -> " + str(c_graph[curr]))
            curr = c_graph[curr]
        P.append(chromosome)
    
    return P

def CycleToChromosome(Nodes):
    chromosome = []
    print(Nodes)
    for i in range(0, len(Nodes), 2):
        if Nodes[i] < Nodes[i+1]:
            chromosome.append(int(Nodes[i+1]/2))
        else:
            chromosome.append(-int(Nodes[i]/2))
            
    return chromosome

for genome in TwoBreakSorting(P, Q):
    out = ""
    for cycle in genome:
        out += "("
        for i in cycle:
            if i > 0:
                out += "+" + str(i) + " "
            else:
                out += str(i) + " "
        out = out.strip(" ") + ")"
    print(out)
```

```python
from collections import defaultdict
from copy import deepcopy

f = open("dataset_865473_3.txt","r")
c = f.readline().split(')')
i1, i2, i3, i4 = [int(x) for x in next(f).split(", ")]
f.close()
P = []
for chromosome in c[:-1]:
    P.append([int(x) for x in chromosome[1:].split()])
    
def TwoBreakOnGenome(P, i1, i2 , i3, i4):
    GenomeGraph = ColoredEdges(P)
    GenomeGraph = TwoBreakOnGenomeGraph(GenomeGraph, i1, i2, i3, i4)
    p = GraphToGenome(GenomeGraph)
    return p

def TwoBreakOnGenomeGraph(GenomeGraph, i1, i2, i3, i4):
    order_flag = 0
    index_i1 = 0
    index_i3 = 0
    
    for edge in range(len(GenomeGraph)):
        if GenomeGraph[edge] == (i1, i2) or GenomeGraph[edge] == (i2, i1):
            if GenomeGraph[edge] == (i2, i1):
                order_flag = 1
            index_i1 = edge
        elif GenomeGraph[edge] == (i3, i4) or GenomeGraph[edge] == (i4, i3):
            index_i3 = edge
    
    if index_i3 < index_i1:
        if not order_flag:
            return GenomeGraph[:index_i3] + [(i4, i2)] + GenomeGraph[index_i1 +1:] + \
                                                [(i1, i3)] + GenomeGraph[index_i3+1:index_i1]
        else:
            return GenomeGraph[:index_i3] + [(i3, i1)] + GenomeGraph[index_i1 +1:] + \
                                                [(i2, i4)] + GenomeGraph[index_i3+1:index_i1]
    else:
        if not order_flag:
            return GenomeGraph[:index_i1] + [(i1, i3)] + GenomeGraph[index_i3 +1:] + \
                                                [(i4, i2)] + GenomeGraph[index_i1+1:index_i3]
        else:
            return GenomeGraph[:index_i1] + [(i2, i4)] + GenomeGraph[index_i3 +1:] + \
                                                [(i1, i3)] + GenomeGraph[index_i1+1:index_i3]
        
def ColoredEdges(P):
    Edges = []
    for chromosome in P:
        Nodes = ChromosomeToCycle(chromosome)
        for i in range(1,len(Nodes) - 2, 2):
            Edges.append((Nodes[i], Nodes[i + 1]))
            
        Edges.append((Nodes[-1], Nodes[0]))
        
    return Edges
    
def ChromosomeToCycle(chromosome):
    Nodes = []
    for i in chromosome:
        if i > 0:
            Nodes.append(2*i - 1)
            Nodes.append(2*i)
        else:
            Nodes.append(-2*i)
            Nodes.append(-2*i - 1)
            
    return Nodes

def GraphToGenome(GenomeGraph):
    P = []
    Cycles = []
    start = defaultdict(int)
    curr_edges = []
    new_cycle_flag = 1
    
    for edge in GenomeGraph:
        if new_cycle_flag:
            new_cycle_flag  = 0
            start[int((edge[0]+1)/2)] = 1
            curr_edges.append(edge)
        else:
            curr_edges.append(edge)
            if start[int((edge[1]+1)/2)]:
                Cycles.append(deepcopy(curr_edges))
                curr_edges = []
                new_cycle_flag = 1
    if len(curr_edges):
        Cycles.append(curr_edges)
    
    Nodes = []
    for cycle in Cycles:
        node_list = []
        for edge in range(len(cycle)-1, 0, -1):
            node_list.append(cycle[edge][1])
            node_list.append(cycle[edge][0])
            
        node_list.append(cycle[0][1])
        node_list.insert(0, cycle[0][0])
        Nodes.append(node_list)
        
    for cycle in Nodes:
        P.append(CycleToChromosome(cycle))
          
    return P

def CycleToChromosome(Nodes):
    chromosome = []
    for i in range(0, len(Nodes), 2):
        if Nodes[i] < Nodes[i+1]:
            chromosome.append(int(Nodes[i+1]/2))
        else:
            chromosome.append(-int(Nodes[i]/2))
            
    return chromosome

out = ""
for cycle in TwoBreakOnGenome(P, i1, i2 , i3, i4):
    out += "("
    for i in cycle:
        if i > 0:
            out += "+" + str(i) + " "
        else:
            out += str(i) + " "
    out = out.strip(" ") + ")"
print(out)
```

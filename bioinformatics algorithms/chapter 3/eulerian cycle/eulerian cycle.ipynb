```python
from collections import defaultdict
from copy import deepcopy

f = open("dataset_865403_2 (1).txt","r")

graph = defaultdict(list)
while line := f.readline():
    node = line.split(':')
    for x in node[1].split():
        graph[int(node[0])].append(int(x))

# hierholzer 
# https://books.google.com/books?id=NvuFAglxaJkC&lpg=PA39&ots=GLw1XLL6rn&dq=%22the%20algorithm%20of%20hierholzer%22&pg=PA39#v=onepage&q&f=false

def EulerianCycle(graph): 
    A = deepcopy(graph) #incidence list
    
    edges_per_vertex = dict()
    for key in A:
        edges_per_vertex[key] = len(A[key])
    
    L = [] #stack
    K = [] #cycle
    
    s = list(A.keys())[0] #first vertex in graph
    L.append(s)
    v = s
    
    while len(L) > 0:
        if edges_per_vertex[v] > 0:
            L.append(v)
            u = A[v][-1]
            edges_per_vertex[v] -= 1
            A[v].pop()
            v = u
        else:
            K.append(v)
            v = L[-1]
            L.pop()
        
    K.reverse()
    return K

out = ""
for v in EulerianCycle(graph):
    out += str(v) + " "
print(out.strip(" "))

f.close()
```

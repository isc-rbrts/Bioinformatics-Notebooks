```python
f = open("dataset_865362_2.txt","r")

pattern = f.readline()
pattern = pattern[:len(pattern) - 1]

def ReverseComplement(pattern):
    complement = ""
    complements = {'A':'T', 'T':'A', 'C':'G', 'G':'C'}
    
    for c in pattern:
        complement = complements[c] + complement
    return complement
    
print(ReverseComplement(pattern))
    
f.close()
```

```python
from collections import defaultdict
from copy import deepcopy

f = open("sample.txt", "r")
g = f.readline().split(')')
f.close()
GenomeGraph = []
for edge in g[:-1]:
    pair = [int(x) for x in edge.split("(")[1].split(", ")]
    GenomeGraph.append((pair[0],pair[1]))
    
def GraphToGenome(GenomeGraph):
    P = []
    Cycles = []
    start = defaultdict(int)
    curr_edges = []
    new_cycle_flag = 1
    
    for edge in GenomeGraph:
        if new_cycle_flag:
            new_cycle_flag  = 0
            start[int((edge[0]+1)/2)] = 1
            curr_edges.append(edge)
        else:
            curr_edges.append(edge)
            if start[int((edge[1]+1)/2)]:
                Cycles.append(deepcopy(curr_edges))
                curr_edges = []
                new_cycle_flag = 1
    
    Nodes = []
    for cycle in Cycles:
        node_list = []
        for edge in range(len(cycle)-1, 0, -1):
            node_list.append(cycle[edge][1])
            node_list.append(cycle[edge][0])
            
        node_list.append(cycle[0][1])
        node_list.insert(0, cycle[0][0])
        Nodes.append(node_list)
        
    for cycle in Nodes:
        P.append(CycleToChromosome(cycle))
          
    return P

def CycleToChromosome(Nodes):
    chromosome = []
    for i in range(0, len(Nodes), 2):
        if Nodes[i] < Nodes[i+1]:
            chromosome.append(int(Nodes[i+1]/2))
        else:
            chromosome.append(-int(Nodes[i]/2))
            
    return chromosome
    
    """
    chromosome = []
    print(Nodes[::-1])
    Nodes = Nodes[::-1]
    for i in range(0, len(Nodes) - 2, 2):
        if Nodes[i] < Nodes[i+1]:
            chromosome.append(int(Nodes[i+1]/2))
        else:
            chromosome.append(-int(Nodes[i]/2))
            
    
    if Nodes[-1] < Nodes[-2]:
        chromosome.insert(0, -int(Nodes[-2]/2))
    else:
        chromosome.insert(0, int(Nodes[-1]/2))
            
    return chromosome
    """

out = ""
for cycle in GraphToGenome(GenomeGraph):
    out += "("
    for i in cycle:
        if i > 0:
            out += "+" + str(i) + " "
        else:
            out += str(i) + " "
    out = out.strip(" ") + ")"
print(out)
    
```

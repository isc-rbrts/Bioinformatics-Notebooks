```python
"""
f = open("","r")

k, t = [int(x) for x in next(f).split()]
Dna = [str(x) for x in next(f).split()]
"""
k, t = 3, 4
Dna = ["GCCCAA", "GGCCTG", "AACCTA", "TTCCTT"]

def GreedyMotifSearch(Dna, k, t):
    BestMotifs = []
    min_loc = 0
    for i in range(len(Dna)):
        BestMotifs.append(Dna[i][:k])
    
    for m in range(len(Dna[0]) - k + 1):
        motifs = [Dna[0][m:m + k]]
        locations = m
        for i in range(1,t):
            
            # Build Profile
            Profile = []
            for j in range(k):
                Profile.append({"A":0.0001,"C":0.0001,"G":0.0001,"T":0.0001})
            for motif in motifs:
                for nucleotide in range(len(motif)):
                    Profile[nucleotide][motif[nucleotide]] += 1.0
            for j in range(k):
                for key in Profile[j]:
                    Profile[j][key] /= i
            # End Build Profile
            motifs.append(ProfileMostProbableKmer(Dna[i], k, Profile))
            for j in range(len(Dna[i]) - k + 1):
                motif = Dna[i][j:j + k]
                if motifs[-1] == motif:
                    locations += j
        
        print(motifs)
        print(Score(motifs), Score(BestMotifs), locations, min_loc)
        if Score(motifs) < Score(BestMotifs):
            BestMotifs = motifs
            min_loc = locations
        elif Score(motifs) == Score(BestMotifs) and locations < min_loc:
            BestMotifs = motifs
            min_loc = locations
            
        
    return BestMotifs
            
            
def ProfileMostProbableKmer(Text, k, Profile):
    MostProbableKmer = ""
    KmerProbability = float('-inf')
    
    for i in range(len(Text) - k + 1):
        kmer = Text[i:i + k]
        prob = 1
        for n in range(len(kmer)):
            prob = prob * Profile[n][kmer[n]]
            
        if prob > KmerProbability:
            KmerProbability = prob
            MostProbableKmer = kmer
    
    return MostProbableKmer

def Score(motifs):
    Profile = []
    k = len(motifs[0])
    for j in range(k):
        Profile.append({"A":0.0,"C":0.0,"G":0.0,"T":0.0})
    for motif in motifs:
        for nucleotide in range(len(motif)):
            Profile[nucleotide][motif[nucleotide]] += 1.0
    
    print(Profile)
    
    score = 0
    for i in range(k):
        max_key = "A"
        max_val = float('-inf')
        for key in Profile[i]:
            if Profile[i][key] > max_val:
                max_val = Profile[i][key]
                max_key = key
        
        for key in Profile[i]:
            if key != max_key:
                score += Profile[i][key]
                
    return score

print(GreedyMotifSearch(Dna, k, t))

                

#f.close()
```

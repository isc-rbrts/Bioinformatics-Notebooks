```python
f = open("dataset_865397_3.txt","r")

k = int(f.readline())
Text = f.readline()
Text = Text[:len(Text) - 1]

def StringComposition(Text, k):
    comp = []
    for i in range(len(Text) - k + 1):
        comp.append(Text[i:i + k])
        
    return comp

out = ""
for kmer in StringComposition(Text, k):
    out += kmer + " "
print(out)

f.close()
```

```python
from collections import defaultdict

f = open("dataset_865532_4.txt","r")
Text = f.readline().strip('\n')
Patterns = [str(x) for x in next(f).split()]
f.close()

C = 5

def MultiplePatternMatching(Text, Patterns):
    Transform = BWT(Text + '$')
    SubstringMatches = BetterBWMatchingPatterns(Transform, Patterns)
    suffix_array = SuffixArray(Text + '$')
    PatternMatches = defaultdict(list)
    
    for i in range(len(Patterns)):
        if SubstringMatches[i] != 0:
            for first_index in range(SubstringMatches[i][0], SubstringMatches[i][1] + 1):
                PatternMatches[Patterns[i]].append(suffix_array[first_index])
        else:
            PatternMatches[Patterns[i]].append(" ")
    
    return PatternMatches  
    
def BWT(Text):
    BWTmatrix = []
    for i in range(len(Text)):
        BWTmatrix.append(Text[i:] + Text[:i])
    
    BWTmatrix.sort()
    
    transformed_text = ""
    for text in BWTmatrix:
        transformed_text += text[-1]
    
    return transformed_text

def BetterBWMatchingPatterns(Transform, Patterns):
    FirstColumn = [str(c) for c in Transform]
    FirstColumn.sort()
    LastColumn = [str(c) for c in Transform]
    
    firstFreqs = defaultdict(int)
    lastFreqs = defaultdict(int)
    
    firstDict = dict()
    lastDict = dict()
    
    Count = defaultdict(list)
    FirstOccurrence = dict()
    
    for i in range(len(LastColumn)):
        first_c = FirstColumn[i]
        last_c = LastColumn[i]
        
        if firstFreqs[first_c] == 0:
            FirstOccurrence.update({first_c: i})
        if lastFreqs[last_c] == 0:
            Count[last_c] = [0] * (i+1)
        
        if len(Count[last_c]) < i + 1:
            Count[last_c] += [Count[last_c][-1]] * (i + 1 - len(Count[last_c]))
        
        firstFreqs[first_c] += 1
        lastFreqs[last_c] += 1
        
        Count[last_c] += [lastFreqs[last_c]]
        
        firstDict.update({(first_c, firstFreqs[first_c]): i})
        lastDict.update({(last_c, lastFreqs[last_c]): i})
        
    for key in Count:
        if len(Count[key]) < len(LastColumn) + 1:
            Count[key] += [Count[key][-1]] * (len(LastColumn) + 1 - len(Count[key]))
    
    SubstringMatches = []
    for Pattern in Patterns:
        SubstringMatches.append(BetterBWMatching(FirstOccurrence, LastColumn, Pattern, Count))
    
    return SubstringMatches

def BetterBWMatching(FirstOccurrence, LastColumn, Pattern, Count):
    top = 0
    bottom = len(LastColumn) - 1
    
    while top <= bottom:
        if len(Pattern):
            symbol = Pattern[-1]
            Pattern = Pattern[:-1]
            foundFlag = 0
            
            for i in range(top, bottom + 1):
                if symbol == LastColumn[i]:
                    top = FirstOccurrence[symbol] + Count[symbol][top]
                    bottom = FirstOccurrence[symbol] + Count[symbol][bottom + 1] - 1
                    foundFlag = 1
                    break
            
            if not foundFlag:
                return 0
            
        else:
            return top, bottom
    
    return 0

def SuffixArray(Text):
    suffixes = []
    for i in range(len(Text)):
        suffixes.append((Text[i:], i))
    
    suffixes.sort()
    
    return [pair[1] for pair in suffixes]

PatternMatches = MultiplePatternMatching(Text, Patterns)
for pattern in PatternMatches:
    out = pattern + ": "
    PatternMatches[pattern].sort()
    for index in PatternMatches[pattern]:
        out += str(index) + " "
    print(out.strip(" "))
```

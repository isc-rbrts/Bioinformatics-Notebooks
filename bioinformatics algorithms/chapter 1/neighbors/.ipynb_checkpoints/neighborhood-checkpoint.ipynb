```python
f = open("dataset_865370_4.txt","r")

pattern = f.readline()
pattern = pattern[:len(pattern) - 1]
d = int(f.readline())

def Neighbors(pattern, d):
    if d == 0:
        return [pattern]
    if len(pattern) == 1:
        return ["A", "C", "T", "G"]
    
    Neighborhood = set()
    SuffixNeighbors = Neighbors(pattern[1:], d)
    
    for string in SuffixNeighbors:
        if HammingDistance(string, pattern[1:]) < d:
            for nucleotide in ["A", "C", "T", "G"]:
                Neighborhood.add((nucleotide + string))
        else:
            Neighborhood.add((pattern[:1] + string))
    return Neighborhood      
        
def HammingDistance(string1, string2):
    if len(string1) != len(string2):
        return -1
    
    dist = 0
    
    for i in range(len(string1)):
        if string1[i] != string2[i]:
            dist +=1
            
    return dist

out = ""
for n in Neighbors(pattern, d):
    out = out + n + " "
print(out[:len(out) - 1])

f.close()
```

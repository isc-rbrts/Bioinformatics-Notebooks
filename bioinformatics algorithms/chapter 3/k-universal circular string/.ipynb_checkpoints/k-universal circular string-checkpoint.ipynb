```python
from collections import defaultdict
from copy import deepcopy
from random import randint

f = open("three.txt","r")

k = int(f.readline())

def KUniversalString(k):
    kmer = ""
    for i in range(k):
        kmer += "0"
    neighborhood = list(Neighbors(kmer, k))
    
    return StringReconstruction(neighborhood, k)[k - 1:]

def Neighbors(pattern, d):
    if d == 0:
        return [pattern]
    if len(pattern) == 1:
        return ["0", "1"]
    
    Neighborhood = set()
    SuffixNeighbors = Neighbors(pattern[1:], d)
    
    for string in SuffixNeighbors:
        if HammingDistance(string, pattern[1:]) < d:
            for nucleotide in ["0", "1"]:
                Neighborhood.add((nucleotide + string))
        else:
            Neighborhood.add((pattern[:1] + string))
    return Neighborhood

def HammingDistance(string1, string2):
    if len(string1) != len(string2):
        return -1
    
    dist = 0
    
    for i in range(len(string1)):
        if string1[i] != string2[i]:
            dist +=1
            
    return dist

def StringReconstruction(Patterns, k):
    graph = DeBrujinString(Patterns)
    path = EulerianPath(graph)
    Text = StringFromGenomePath(path)
    return Text
    
def DeBrujinString(Patterns):
    graph = defaultdict(lambda: defaultdict(lambda: 0))
    for i in range(len(Patterns)):
        graph[Patterns[i][:len(Patterns[i]) - 1]][Patterns[i][1:]] += 1
    
    reduced_graph = defaultdict(list)
    for node1 in graph:
        for node2 in graph[node1]:
            for i in range(graph[node1][node2]):
                reduced_graph[node1].append(node2)
    
    return reduced_graph

def StringFromGenomePath(path):
    Text = path[0]
    for i in range(1, len(path)):
        Text += path[i][-1]
        
    return Text

def EulerianPath(graph): 
    s = list(graph.keys())[0]
    end = FindEnd(graph, s)
    graph_r = ReverseGraph(graph)
    return FindPath(graph_r, end)

def FindEnd(graph, start): 
    A = deepcopy(graph) #incidence list
    
    edges_per_vertex = dict()
    for key in A:
        edges_per_vertex[key] = len(A[key])
    
    L = [] #stack
    K = [] #cycle
    
    s = start
    L.append(s)
    v = s
    
    while len(L) > 0:
        if v not in edges_per_vertex:
            return v
        
        if edges_per_vertex[v] > 0:
            i = randint(0, len(A[v]) - 1)
            L.append(v)
            u = A[v][i]
            edges_per_vertex[v] -= 1
            A[v].pop(i)
            v = u
        else:
            K.append(v)
            v = L[-1]
            L.pop()
    
    return K[0]
        
def FindPath(graph, end):
    A = deepcopy(graph) #incidence list
    
    edges_per_vertex = dict()
    for key in A:
        edges_per_vertex[key] = len(A[key])
    
    L = [] #stack
    K = [] #cycle
    
    s = end
    L.append(s)
    v = s
    start_flag = 0
    
    while len(L) > 0:
        
        if v not in edges_per_vertex:
            start_flag = 1
            start = v
            A[v].append(end)
            edges_per_vertex[v] = 0
            edges_per_vertex[v] += 1
        
        if edges_per_vertex[v] > 0:
            L.append(v)
            u = A[v][-1]
            edges_per_vertex[v] -= 1
            A[v].pop()
            v = u
        else:
            K.append(v)
            v = L[-1]
            L.pop()
    
    return K
    

def ReverseGraph(graph):
    graph_r = defaultdict(list)
    for key in graph:
        for i in graph[key]:
            graph_r[i].append(key)
            
    return graph_r

print(KUniversalString(k))

f.close()
```

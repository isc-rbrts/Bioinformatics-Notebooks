```python
from collections import defaultdict

f = open("dataset_865399_6.txt","r")

k = int(f.readline())
Text = f.readline()
Text = Text[:len(Text) - 1]

def DeBrujinString(Text, k):
    Patterns = [Text[i:i + k] for i in range(len(Text) - k + 1)]
    
    graph = defaultdict(lambda: defaultdict(lambda: 0))
    for i in range(len(Patterns)):
        graph[Patterns[i][:len(Patterns[i]) - 1]][Patterns[i][1:]] += 1
    
    reduced_graph = defaultdict(list)
    for node1 in graph:
        for node2 in graph[node1]:
            for i in range(graph[node1][node2]):
                reduced_graph[node1].append(node2)
    
    return reduced_graph

g = DeBrujinString(Text, k)
for key in g:
    out = key + ": "
    for node in g[key]:
        out += node + " "
    print(out)

f.close()
```

```python
f = open("dataset_865536_2.txt","r")
Text = f.readline().strip('\n')
K = int(f.readline().strip('\n'))
f.close()

def PartialSuffixArray(Text, K):
    suffixes = []
    for i in range(len(Text)):
        suffixes.append((Text[i:], i))
    
    suffixes.sort()
    
    SuffixArray = [pair[1] for pair in suffixes]
    
    PartialSuffixArray = []
    for i in range(len(SuffixArray)):
        if SuffixArray[i] % K == 0:
            PartialSuffixArray.append((i, SuffixArray[i]))
        
    return PartialSuffixArray

for item in PartialSuffixArray(Text, K):
    print(item[0],item[1])
```

```python

f = open("dataset_865383_9.txt","r")

k, t = [int(x) for x in next(f).split()]
Dna = [str(x) for x in next(f).split()]

def GreedyMotifSearch(Dna, k, t):
    BestMotifs = []
    for i in range(len(Dna)):
        BestMotifs.append(Dna[i][:k])
    
    for m in range(len(Dna[0]) - k + 1):
        motifs = [Dna[0][m:m + k]]
        for i in range(1,t):
            
            # Build Profile
            Profile = []
            for j in range(k):
                Profile.append({"A":1.0,"C":1.0,"G":1.0,"T":1.0}) # 0.0 for no pseudocounts
            for motif in motifs:
                for nucleotide in range(len(motif)):
                    Profile[nucleotide][motif[nucleotide]] += 1.0
            for j in range(k):
                for key in Profile[j]:
                    Profile[j][key] /= (i + 4) # without adding 4 for no pseudocounts
            # End Build Profile
            motifs.append(ProfileMostProbableKmer(Dna[i], k, Profile))
        
        if Score(motifs) < Score(BestMotifs):
            BestMotifs = motifs
            
        
    return BestMotifs
            
            
def ProfileMostProbableKmer(Text, k, Profile):
    MostProbableKmer = ""
    KmerProbability = float('-inf')
    
    for i in range(len(Text) - k + 1):
        kmer = Text[i:i + k]
        prob = 1
        for n in range(len(kmer)):
            prob = prob * Profile[n][kmer[n]]
            
        if prob > KmerProbability:
            KmerProbability = prob
            MostProbableKmer = kmer
    
    return MostProbableKmer

def Score(motifs):
    Profile = []
    k = len(motifs[0])
    for j in range(k):
        Profile.append({"A":0.0,"C":0.0,"G":0.0,"T":0.0})
    for motif in motifs:
        for nucleotide in range(len(motif)):
            Profile[nucleotide][motif[nucleotide]] += 1.0
    
    score = 0
    for i in range(k):
        max_key = "A"
        max_val = float('-inf')
        for key in Profile[i]:
            if Profile[i][key] > max_val:
                max_val = Profile[i][key]
                max_key = key
        
        for key in Profile[i]:
            if key != max_key:
                score += Profile[i][key]
                
    return score


out = ""
for motif in GreedyMotifSearch(Dna, k, t):
    out = out + motif + " "
print(out)

                

f.close()
```

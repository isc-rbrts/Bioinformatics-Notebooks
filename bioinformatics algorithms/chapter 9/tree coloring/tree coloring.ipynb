```python
from collections import defaultdict

f = open("dataset_865535_6.txt", "r")
graph = defaultdict(list)
while (line := f.readline().strip('\n')) != "-":
    split_line = line.split(':')
    for toNode in [int(x) for x in split_line[1].split()]:
        graph[int(split_line[0])].append(toNode)

colors = defaultdict(lambda: "gray")
while line := f.readline().strip('\n'):
    split_line = [str(x) for x in line.split()]
    colors[int(split_line[0])] = split_line[1]
    
f.close()

def TreeColoring(graph, colors):
    L = TopologicalOrdering(graph)[::-1]
    
    while len(L):
        curr = L[0]
        if colors[curr] == "gray":
            child_colors = set()
            for child in graph[curr]:
                child_colors.add(colors[child])
                
            if len(child_colors) == 1:
                colors[curr] = list(child_colors)[0]
            else:
                colors[curr] = "purple"
            
        L.pop(0)
        
    return colors

def TopologicalOrdering(graph):
    reverse_graph = defaultdict(list)
    for key in graph:
        for node in graph[key]:
            reverse_graph[node].append(key)
    
    L = list()
    Candidates = set()
    for key in graph:
        if key not in reverse_graph:
            Candidates.add(key)
    while len(Candidates):
        a = list(Candidates)[0]
        L.append(a)
        Candidates.remove(a)
        
        for node in graph[a]:
            if a in reverse_graph[node]:
                reverse_graph[node].remove(a)
                
            if len(reverse_graph[node]) == 0:
                Candidates.add(node)
                
    for key in reverse_graph:
        if len(reverse_graph[key]):
            return "Not a DAG"
        
    return L

ColoredTree = TreeColoring(graph, colors)
for node in ColoredTree:
    print(str(node) + " " + str(ColoredTree[node]))
```
